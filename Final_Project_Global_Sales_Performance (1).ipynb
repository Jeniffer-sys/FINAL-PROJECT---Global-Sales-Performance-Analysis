# üè∑Ô∏è Attribute Information :

The dataset contains 11 attributes (columns). Each attribute is described below along with its significance for sales and financial analysis:

1. **Region**  
   - Type: Categorical (string)
   - Description: Geographical area where the sales were made (e.g., East, West, North, South).  
   - Relevance: Helps analyze regional performance, compare sales territories, and optimize resource allocation.

2. **Salesperson**  
   - Type: Categorical (string)  
   - Description: Name or identifier of the salesperson responsible for the sales..  
   - Relevance: Useful for tracking individual performance, incentive planning, and identifying training needs..

3. **Revenue**  
   - Type: Numerical (integer)  
   - Description: Total income generated from sales before deducting expenses.  
   - Relevance: Key measure of sales performance, supports forecasting, and helps evaluate sales growth..

4. **Profit**  
   - Type: Numerical (integer)  
   - Description: Net gain after deducting costs from revenue.  
   - Relevance: Indicates business profitability, helps identify high- vs. low-margin products/regions, and supports pricing decisions.

5. **Units_Sold**  
   - Type: Numerical (integer)  
   - Description: Number of product units sold.  
   - Relevance: Helps measure sales volume, understand demand trends, and plan inventory.

6. **Customer_Satisfaction**  
   - Type: Numerical (float)  
   - Description: Customer rating or satisfaction score (e.g., scale of 1‚Äì5 or 1‚Äì10).  
   - Relevance: Reflects customer experience, helps identify areas of improvement, and correlates with customer retention.

7. **Marketing_Spend**  
   - Type: Numerical (integer)  
   - Description: Money spent on marketing activities to drive sales.  
   - Relevance: Allows calculation of return on marketing investment (ROMI), measures campaign effectiveness, and links spend to sales outcomes.

8. **Discount_Percentage**  
   - Type: Numerical (float/percentage)  
   - Description: Discount applied on sales price (in percentage).  
   - Relevance: Influences sales volume and profit margins; helps assess discount strategies.

9. **Product_Category**  
   - Type: Categorical (string)  
   - Description: Classification of products (e.g., Electronics, Furniture, Clothing).  
   - Relevance: Enables product portfolio analysis, identifies best-selling categories, and guides strategic investments.

10. **Sales_Channel**  
    - Type: Categorical (string)  
    - Description: Channel through which sales were made (e.g., Online, Retail, Distributor).  
    - Relevance: Helps compare performance across channels, evaluate effectiveness, and plan distribution strategy.

11. **Order_Processing_Time**  
    - Type: Numerical (integer)  
    - Description: Time taken to process an order (likely in days).  
    - Relevance: Indicates operational efficiency, impacts customer satisfaction, and helps optimize supply chain performance.

# üßπB. Data Pre-processing

In this step, we clean and prepare the dataset for analysis. The following tasks are performed:

- **Handling missing values** ‚Üí Check for missing data and decide whether to fill or remove them.  
- **Removing duplicates** ‚Üí Drop duplicate records to ensure data accuracy.  
- **Correcting data types** ‚Üí Convert columns to appropriate types (e.g., numerical, categorical, datetime).  
- **Creating derived columns** ‚Üí Generate new useful columns from existing ones.  
- **Filtering or aggregating data** ‚Üí Remove unnecessary rows or summarize data for analysis.  

## üß© 1. Handling missing values

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# Example: Fill numeric missing values with mean (if any)

df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df
```

```python
# Fill categorical missing values with mode (fixed version)

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

## üóëÔ∏è 2. Removing duplicates

```python
print("\nNumber of duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
df
```

## üî§ 3. Correcting data types

```python
#Example: Convert 'Order_Processing_Time' to categorical if needed

df['Order_Processing_Time'] = df['Order_Processing_Time'].astype('category')
```

```python
df
```

## üßÆ 4. Creating derived columns

```python
# Example: Create Profit Margin column

df['Profit_Margin'] = (df['Profit'] / df['Revenue']) * 100
```

```python
df
```

```python
# Example: Create Revenue per Unit column

df['Revenue_per_Unit'] = df['Revenue'] / df['Units_Sold']
```

```python
df
```

# üîç C.Exploratory Data Analysis (EDA)
  ## Conduct descriptive and exploratory analysis to uncover patterns and trends:

## Univariate Analysis

```python
# Revenue distribution
sns.histplot(df["Revenue"], bins=30, kde=True, color="skyblue")
plt.title("Revenue Distribution")
plt.show()
```

```python
# Profit distribution
sns.histplot(df["Profit"], bins=30, kde=True, color="orange")
plt.title("Profit Distribution")
plt.show()
```

## Bivariate Analysis

```python
# Revenue vs Profit
sns.scatterplot(x="Revenue", y="Profit", hue="Region", data=df, alpha=0.7)
plt.title("Revenue vs Profit by Region")
plt.show()
```

```python
# Discount vs Profit
sns.scatterplot(x="Discount_Percentage", y="Profit", hue="Product_Category", data=df, alpha=0.7)
plt.title("Discount % vs Profit by Product Category")
plt.show()
```

## Multivariate Analysis (Groupby & Pivot Tables)

```python
# Pivot: Region √ó Product Category
pivot_region_category = pd.pivot_table(
    df, values=["Revenue","Profit"],
    index="Region", columns="Product_Category", aggfunc="mean"
)
pivot_region_category
```

## Correlation Analysis

```python
# Correlation heatmap
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

## Bar plot - Revenue by Region

```python
plt.figure(figsize=(8,6))
sns.barplot(
    x="Region",
    y="Revenue",
    data=df,
    estimator=sum,
    errorbar=None,          # replaces ci=None
    hue="Region",           # needed when using palette
    palette="viridis",
    legend=False            # avoid duplicate legends
)
plt.title("Total Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show()
```
